```python
# Usaremos apenas o módulo 're' para expressões regulares
import re
import regex
import unicodedata
```

## Função para extração de e-mails

Esta função recebe um texto e retorna uma lista de e-mails válidos.
Um e-mail é considerado válido se contém '@' e termina com .com, .org, .net ou .br

```python
def extract_emails(text):
    """
    Recebe uma string de texto e retorna uma lista de e-mails válidos.
    Um e-mail é válido se contém '@' e termina com .com, .org, .net ou .br
    """
    pattern = r'\b[\w\.-]+@[\w\.-]+\.(?:com|org|net|br)\b'
    emails = re.findall(pattern, text, re.IGNORECASE)
    return emails
```

## Função para validação de IDs

Esta função recebe uma lista de IDs e separa em válidos e inválidos.
Um ID válido segue o padrão AAA-1234 (3 letras maiúsculas, hífen, 4 dígitos)

```python
def validate_ids(id_list):
    """
    Recebe uma lista de IDs e retorna duas listas:
    - IDs válidos (AAA-1234)
    - IDs inválidos
    """
    valid = []
    invalid = []
    pattern = r'^[A-Z]{3}-\d{4}$'
    for id in id_list:
        if re.match(pattern, id):
            valid.append(id)
        else:
            invalid.append(id)
    return valid, invalid
```

## Classe TextCleaner

Esta classe permite limpar textos, removendo espaços extras, caracteres especiais
(mantendo apenas letras, números, espaços, pontos e vírgulas) e convertendo para minúsculas.

```python
class TextCleaner:
    def __init__(self, original_text):
        self.original_text = original_text
        self.cleaned_text = original_text

    def remove_extra_spaces(self):
        self.cleaned_text = re.sub(r'\s+', ' ', self.cleaned_text).strip()
        return self

    def remove_special_characters(self):
        self.cleaned_text = re.sub(
            r'[^A-Za-z0-9\s\.,@_\-\u00C0-\u024F]',
            '',
            self.cleaned_text
        )
        return self

    def remove_accents(self):
        nfkd = unicodedata.normalize('NFKD', self.cleaned_text)
        self.cleaned_text = ''.join(
            c for c in nfkd if not unicodedata.combining(c)
        )
        return self

    def to_lowercase(self):
        self.cleaned_text = self.cleaned_text.lower()
        return self

    def get_text(self):
        return self.cleaned_text
```

## Função para ler arquivo com tratamento de erro

Esta função tenta ler um arquivo .txt e retorna o conteúdo como string.
Se houver erro (arquivo não encontrado, codificação inválida, etc.), retorna uma mensagem de erro.

```python
def read_file(filename):
    """
    Tenta ler um arquivo .txt e retorna o conteúdo como string.
    Em caso de erro, retorna uma mensagem de erro.
    """
    try:
        with open(filename, 'r', encoding='utf-8') as f:
            return f.read()
    except FileNotFoundError:
        return "Arquivo não encontrado!"      
```

## Função para contar frequência de palavras

Esta função recebe um texto e retorna um dicionário com a frequência de cada palavra,
ignorando maiúsculas/minúsculas e pontuação.

```python
def word_frequency(text):
    """
    Recebe uma string de texto e retorna um dicionário {palavra: contagem}
    Ignora maiúsculas/minúsculas e pontuação.
    """
    # Remove pontuação
    text = re.sub(r'[^\w\s]', '', text)
    words = text.lower().split()
    freq = {}
    for word in words:
        freq[word] = freq.get(word, 0) + 1
    return freq
```
